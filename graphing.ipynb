```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import json
```

```python
df = pd.read_csv("nolist.csv").T.iloc[1:]
colList = ["LR_acc", "LR_f1", "LR_r2",
           "NB_acc", "NB_f1", "NB_r2",
           "vader_acc", "vader_f1", "vader_r2",
           "ord_acc", "ord_f1", "ord_r2",
           "LAT_acc", "LAT_f1", "LAT_r2",
           "LIT_acc", "LIT_f1", "LIT_r2",
           "LSE_acc", "LSE_f1", "LSE_r2",
          ]
df.columns = colList

df
```

```python
# accuracy
#df[df < 0] = 0
ax = df[["LR_acc", "NB_acc", "vader_acc", "ord_acc", "LAT_acc", "LIT_acc", "LSE_acc"]].plot.bar(title = "Accuracy", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Accuracy")
ax.set_xlabel("Category")
```

```python
#df[df < 0] = 0
ax = df[["LR_f1", "NB_f1", "vader_f1", "ord_f1", "LAT_f1", "LIT_f1", "LSE_f1"]].plot.bar(title = "F1 Score", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("F1 Score")
ax.set_xlabel("Category")
```

```python
#df[df < 0] = 0
ax = df[["LR_r2", "NB_r2", "vader_r2", "ord_r2", "LAT_r2", "LIT_r2", "LSE_r2"]].plot.bar(title = "R squared", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("R squared")
ax.set_xlabel("Category")
```
