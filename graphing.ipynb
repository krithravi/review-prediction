```python
import pandas as pd
import matplotlib as plt
```

```python
results_df = pd.read_csv("lin373n.csv")
results_df.head()
```

```python
ax = results_df.plot.bar(title = "Accuracy", x = "Category", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Percent correct")
ax.set_xlabel("Category")
```
