```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors
import json
```

```python
df = pd.read_csv("nolist.csv").T.iloc[1:]
colList = ["LR_acc", "LR_f1", "LR_r2",
           "NB_acc", "NB_f1", "NB_r2",
           "vader_acc", "vader_f1", "vader_r2",
           "ord_acc", "ord_f1", "ord_r2",
           "LAT_acc", "LAT_f1", "LAT_r2",
           "LIT_acc", "LIT_f1", "LIT_r2",
           "LSE_acc", "LSE_f1", "LSE_r2",
          ]
df.columns = colList

df
```

```python
# accuracy
#df[df < 0] = 0
#ax = df[["LR_acc", "NB_acc", "vader_acc", "ord_acc", "LAT_acc", "LIT_acc", "LSE_acc"]].plot.bar(title = "Accuracy", figsize=(20,10))
mynewDF = df[["LR_acc", "NB_acc", "vader_acc"]]
mynewDF.columns = ["LR", "NB", "Vader"]
ax = mynewDF.plot.bar(title = "Prediction Accuracy in Product Categories", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(0.8, 0.9));
ax.set_ylabel("Accuracy")
ax.set_xlabel("Category")
```

```python
#df[df < 0] = 0
#ax = df[["LR_f1", "NB_f1", "vader_f1", "ord_f1", "LAT_f1", "LIT_f1", "LSE_f1"]].plot.bar(title = "F1 Score", figsize=(20,10))
mynewDF = df[["LR_f1", "NB_f1", "vader_f1"]]
mynewDF.columns = ["LR", "NB", "Vader"]
ax = mynewDF.plot.bar(title = "F1 Score in Product Categories", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(0.8, 0.9));
ax.set_ylabel("F1 Score")
ax.set_xlabel("Category")
```

```python
#df[df < 0] = 0
#ax = df[["LR_r2", "NB_r2", "vader_r2", "ord_r2", "LAT_r2", "LIT_r2", "LSE_r2"]].plot.bar(title = "R squared", figsize=(20,10))
mynewDF = df[["LR_r2", "NB_r2", "vader_r2"]]
mynewDF.columns = ["LR", "NB", "Vader"]
ax = mynewDF.plot.bar(title = "R Squared Value in Product Categories", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(0.9, 0.2));
ax.set_ylabel("R squared")
ax.set_xlabel("Category")
```

```python
myDF = pd.read_csv("offBy.csv", index_col=0)
myDF
```

```python
mynewDF = myDF[["LR0", "LR1", "LR2", "LR3", "LR4"]]
mynewDF.columns = ["0", "1", "2", "3", "4"]
ax = mynewDF.plot.bar(title = "Prediction Distance for Logistic Regression", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Proportion of number of stars prediction off by")
ax.set_xlabel("Category")
```

```python
mynewDF = myDF[["NB0", "NB1", "NB2", "NB3", "NB4"]]
mynewDF.columns = ["0", "1", "2", "3", "4"]

ax = mynewDF.plot.bar(title = "Prediction Distance for Naive Bayes", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Proportion of number of stars prediction off by")
ax.set_xlabel("Category")
```

```python
mynewDF = myDF[["Vad0", "Vad1", "Vad2", "Vad3", "Vad4"]]
mynewDF.columns = ["0", "1", "2", "3", "4"]

ax = mynewDF.plot.bar(title = "Prediction Distance for Vader", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Proportion of number of stars prediction off by")
ax.set_xlabel("Category")
```

```python
mynewDF = myDF[["LR0", "NB0", "LR1", "NB1", "LR2", "NB2", "LR3", "NB3", "LR4", "NB4"]]
mynewDF.columns = ["0-LR", "0-NB", "1-LR", "1-NB", "2-LR", "2-NB", "3-LR", "3-NB", "4-LR", "4-NB"]

#colors = ["#0000FF", "#4D4DFF", "#FF8817", "#FFB817", "#"]
colors = ["blue", "cornflowerblue", "green", "limegreen", "red", "salmon", "purple", "orchid", "darkorange", "orange"]

ax = mynewDF.plot.bar(title = "Prediction Distance Comparison, Logistic Regression and Naive Bayes", figsize=(20,10), color=colors)
ax.legend(loc='center left', bbox_to_anchor=(1.0, 0.5));
ax.set_ylabel("Proportion of number of stars prediction off by")
ax.set_xlabel("Category")
```

```python
mynewDF = myDF.copy()
mynewDF.loc["Average"] = myDF.mean(axis=0)
avgDF = pd.DataFrame()
avgDF["LR"] = list(mynewDF.loc["Average"][["LR0", "LR1", "LR2", "LR3", "LR4"]])
avgDF["NB"] = list(mynewDF.loc["Average"][["NB0", "NB1", "NB2", "NB3", "NB4"]])
avgDF["Vad"] = list(mynewDF.loc["Average"][["Vad0", "Vad1", "Vad2", "Vad3", "Vad4"]])
ax = avgDF.plot.bar(title = "Average Prediction Distance Across Categories", figsize=(20,10))
ax.legend(loc='center left', bbox_to_anchor=(0.9, 0.9));
ax.set_ylabel("Proportion of number of stars prediction off by")
ax.set_xlabel("Stars off by")
plt.xticks(rotation=0)
plt.show()
```
